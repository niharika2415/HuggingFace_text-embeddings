```python
import tensorflow as tf
from transformers import AutoTokenizer, TFAutoModel
import numpy as np
```

```python
tokenizer= AutoTokenizer.from_pretrained("openai/clip-vit-base-patch32")
model= TFAutoModel.from_pretrained("openai/clip-vit-base-patch32")
```

```python
texts=[
    "A red car on a rainy day",
    "Show me a gray sky",
    "I want to go to the beach",
    "I want to wear a black dress"
]
```

```python
inputs= tokenizer(
 texts,
 padding= True,
 truncation=True,
 return_tensors="tf"
)
```

```python
print(inputs['input_ids'].shape) # (batch size, seq len)
```

```python
text_embeddings = model.get_text_features(input_ids=inputs["input_ids"])
```

```python
print("Embeddings shape:", text_embeddings.shape) # (batch_size, hidden_size)
```

```python
embeddings_np= text_embeddings.numpy()
```

```python
# Step 6: Save embeddings
import pickle
# Path to your folder
save_dir = "/content/drive/MyDrive/05-text-image-embedding"

# Convert embeddings to numpy
embeddings_np = text_embeddings.numpy()

# 1. Save as NumPy .npy
np.save(f"{save_dir}/embeddings.npy", embeddings_np)

# 2. Save as CSV
np.savetxt(f"{save_dir}/embeddings.csv", embeddings_np, delimiter=",")

# 3. Save as Pickle
with open(f"{save_dir}/embeddings.pkl", "wb") as f:
    pickle.dump(embeddings_np, f)

print(" Embeddings saved in all formats at:", save_dir)
```
